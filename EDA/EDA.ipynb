# EDA Technique Lets Applay 
- Missing Value Tratment For Numerical Data

# 1. Missing Value Tratment

```python
cleand_data
```

```python
cleand_data.isnull().sum()
```

#### 1. Numerical Missing

```python
cleand_data
```

```python
cleand_data['Age'] = cleand_data['Age'].fillna(np.mean(pd.to_numeric(cleand_data['Age'])))
```

```python
cleand_data['Age']
```

```python
cleand_data['Exp'] = cleand_data['Exp'].fillna(np.mean(pd.to_numeric(cleand_data['Exp'])))
```

```python
cleand_data['Exp']
```

```python
cleand_data
```

#### 2. Catagorical Missing

```python
cleand_data['Location']
```

```python
cleand_data['Location'] = cleand_data['Location'].fillna(cleand_data['Location'].mode()[0])
```

```python
cleand_data['Location']
```

```python
cleand_data
```

```python
cleand_data.info()
```

```python
cleand_data['Age'] = cleand_data['Age'].astype(int)
```

```python
cleand_data['Salary'] = cleand_data['Salary'].astype(int)
```

```python
cleand_data['Exp'] = cleand_data['Exp'].astype(int)
```

```python
cleand_data["Name"] = cleand_data["Name"].astype('category')
cleand_data["Domain"] = cleand_data["Domain"].astype('category')
cleand_data["Location"] = cleand_data["Location"].astype('category')
```

```python
cleand_data.info()
```

```python
cleand_data
```

```python
cleand_data.to_csv("Clean_data.csv")
```
