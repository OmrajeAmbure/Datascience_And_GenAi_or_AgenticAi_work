````python
def clean_address(text):
    # If the row is truly missing (NaN), keep it as NaN
    if pd.isna(text):
        return text
        
    # Keep letters, numbers, spaces, AND commas
    cleaned = re.sub(r'[^a-zA-Z0-9, ]', '', str(text))
    
    # Clean up any accidental double spaces or double commas
    cleaned = re.sub(r'\s+', ' ', cleaned)
    cleaned = re.sub(r',+', ',', cleaned)
    
    return cleaned.strip()
````

````python
import ast
def clean_reviews(review_string):
    # Handle empty rows or empty lists
    if pd.isna(review_string) or review_string == '[]':
        return []
    try:
        # 1. Convert the text string into an actual Python list of tuples
        actual_list = ast.literal_eval(review_string)
        
        cleaned_list = []
        for rating_text, review_text in actual_list:
            # 2. Extract just the rating number (e.g., 'Rated 4.0' -> 4.0)
            rating_num = re.search(r'[0-9.]+', rating_text)
            rating = float(rating_num.group()) if rating_num else None
            
            # 3. Clean up the review body text
            # Remove 'RATED\n', replace newlines/tabs with spaces
            review_text = review_text.replace('RATED\n', '').replace('\n', ' ').replace('\r', ' ')
            
            # 4. Fix Mojibake/corrupted characters (e.g., converting 'isnÃƒ...' back to "isn't")
            # review_text = ftfy.fix_text(review_text)
            # Use this built-in Python alternative:
            review_text = review_text.encode('raw_unicode_escape').decode('utf-8', errors='ignore')

            # 5. Clean up multiple spaces
            review_text = re.sub(r'\s+', ' ', review_text).strip()
            
            cleaned_list.append((rating, review_text))
            
        return cleaned_list
        
    except Exception as e:
        # If a row has broken brackets and can't be parsed, return it empty
        return []

# Apply the function to the entire dataset
df['cleaned_reviews'] = df['reviews_list'].apply(clean_reviews)
````

| Skewness Value | Meaning                 |
| -------------: | ----------------------- |
|              0 | Perfectly symmetric     |
|    -0.5 to 0.5 | Approximately symmetric |
|       0.5 to 1 | Moderately right skewed |
|            > 1 | Highly right skewed     |
|     -0.5 to -1 | Moderately left skewed  |
|           < -1 | Highly left skewed      |

**Why is Skewness Important?**
1. Choosing Mean or Median
- This is the biggest practical use.
- Suppose your salary data is:
    30
    35
    40
    40
    45
    50
    500
- Mean: 105
- Median:40

**Which represents a typical salary?**
- Median, because the ₹500 salary is an outlier.
- Rule:
    - Symmetric data → Mean is appropriate.
    - Skewed data → Median is preferred.
