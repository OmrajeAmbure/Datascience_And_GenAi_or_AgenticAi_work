### 4) Write a program where the user is asked to enter two integers (divisor and dividend) and the quotient and the remainder of their division is computed.(Both divisor and dividend should be integers.)

```python
print("Enter The Two Number (divisor and dividend): ")
try:
    divisor = int(input())
    dividend = int(input())
    if divisor == 0:
        print("Error: Division by zero is not allowed.")
    else:
        quotient = dividend//divisor;
        reminder = dividend%divisor;

        print(f"Quotient: {quotient}")
        print(f"Remainder: {reminder}")
except ValueError:
    print("Invalid Input! Please Enter Valid Integers. ")
```

### 6) Write a Program to Swap Two Numbers

```python
# using Third variable
print("Enter The Two Number To Perform Swapping : ")
num1 = int(input())
num2 = int(input())
print("Before Swapping: ");
print("Number 1: ",num1,"\nNumber 2: ",num2);

temp = num1
num1 = num2
num2 = temp
print("After Swapping: ")
print("Number 1: ",num1,"\nNumber 2: ",num2);
```

```python
# without using third variable
print("Enter The Two Number To Perform Swapping : ")
num1 = int(input())
num2 = int(input())
print("Before Swapping: ");
print("Number 1: ",num1,"\nNumber 2: ",num2);
# using Third variable
num1,num2 = num2,num1
print("After Swapping: ")
print("Number 1: ",num1,"\nNumber 2: ",num2);
```

### 1) Write a Program to Check Whether Number is Even or Odd

```python
print("Enter The Number :")
num = int(input());

if num%2==0:
    print('Even Number');
elif num%2==1:
    print("Odd Number");
```

### 2) Write a Program to Check Whether a Character is Vowel or Consonant.

```python
print("Enter The Character : ")
char_str = input();
if char_str == 'a' or char_str == 'e' or char_str == 'i' or char_str == 'o' or char_str == 'u':
    print(f"The Given Character {char_str} is Vowel");
else:
    print(f"The Given Character {char_str} is Consonant")
```

### 3) Write a Program to Find Largest Number Among Three Numbers entered by users


```python
print("Enter The Three Number To Find Largest Number Among Three Numbers : ")
num1 = int(input())
num2 = int(input())
num3 = int(input())

if num1 > num2 and num1 > num3:
    print(f"{num1} is Largest Number")
elif num2>num1 and num2>num3:
    print(f"{num2} is Largest Number")
else:
    print(f"{num3} is Largest Number")
```
