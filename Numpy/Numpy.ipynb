# Array Manapulation Functions

```python
from numpy import *

# Reshape an array
a1 = np.array([1,2,3,4]) 
a1
```

```python
# Flatten an array
f1 = np.array([
    [1,2],
    [3,4],
    [5,6],
    [7,8]
])
flattened = np.ravel(f1) # Return a contiguous flattened array. 1D array
print("Flattened array : ",flattened)
```

```python
e1 = np.array([[1,2],[3,4]])
transposed = np.transpose(e1) # it convert row into columan or columan into row
transposed
# For a 1-D array, this returns an unchanged view of the original array, as a
# transposed vector is simply the same vector.
```

```python
a1 = np.array([1,2])
b1 = np.array([3,4])
a2 = np.array([1,2])
b2 = np.array([3,4])
stacked = np.vstack([a1,b1,a2,b2]) # Stack arrays in sequence vertically (row wise) 
stacked
```

```python
a1 = np.array([1,2])
b1 = np.array([3,4])
stacked = np.hstack([a1,b1]) # Stack arrays in sequence horizontally (column wise).
stacked
```
