## BitWise Operator

1.Complement(~) --> 

- n = 2
- Binary form of 2 = 0010
- Bitwise complement operation on 2 = ~ 0010 
                                                         = 1101 + 1 ==> 0111
- 1101 is equivalent to decimal value 13.

    - Expected output: 13
    - Correct Output : -3

The compiler returns the 2's complement of the input value.


```python
~2
```

```python
~13
```

```python
6 & 4 """ -->    0110 
                 0100
                ------ 
                 0100 --> 4
"""        
```

```python
6 | 4 """ -->  0110
               0100
              ------
               0110 --> 6
"""
```
