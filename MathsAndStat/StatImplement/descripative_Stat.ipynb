```python
# 1. Create the crosstab
month_exp_tmp = pd.crosstab(index=income_df['Mthly_HH_Expense'], columns='count')
month_exp_tmp.reset_index(inplace=True)

# 2. Find the Mthly_HH_Expense with the highest count (the mode)
max_val = month_exp_tmp['count'].max()
most_frequent_row = month_exp_tmp[month_exp_tmp['count'] == max_val]

print(most_frequent_row)
```
